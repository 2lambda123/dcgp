```python
import dcgpy
%matplotlib inline
```

```python
nonlinearities = dcgpy.kernel_set_double(["sig", "ReLu", "tanh"])
dcgpann = dcgpy.expression_ann_double(inputs=1, outputs=1, rows=10, cols=6, 
                                      levels_back=2, arity=10, kernels=nonlinearities(), seed=32)
dcgpann.randomise_weights(seed=45)
dcgpann.visualize(show_nonlinearities=False)
```

```python
dcgpann.mse([[1],[2.3],[-0.3]],[[0.2],[0.4],[0.5]])
```

```python
dcgpann([0.01])
```
